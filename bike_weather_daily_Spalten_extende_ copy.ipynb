```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("CBS_2021-2023_Daily_Weather.csv", sep =',')

df.head()
```

```python
#Hinzufügen der Spalte Werktage
import pandas as pd

df["Date"] = pd.to_datetime(df["Date"])

# Prüfen, ob der Tag ein Wochenende ist (Samstag=5, Sonntag=6)
df["is_weekend"] = df["Date"].dt.weekday >= 5  # True = Wochenende, False = Werktag

# Optional: Als Text anzeigen
df["day_type"] = df["is_weekend"].map({True: "Weekend", False: "Workday"})

df.head()
```

```python
#Hinzufügen der Spalte Holiday

import holidays
# Sicherstellen, dass die Date-Spalte datetime ist
df["Date"] = pd.to_datetime(df["Date"])

# US-Feiertage abrufen (ändere "US" zu einem anderen Ländercode, falls nötig)
us_holidays = holidays.US()

# Prüfen, ob das Datum ein Feiertag ist
df["is_holiday"] = df["Date"].apply(lambda x: "Yes" if x in us_holidays else "No")
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
#Hinzufüger einer Spalte für Seasons

# Funktion zur Bestimmung der Jahreszeit basierend auf dem Monat
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Fall"

# Neue Spalte "season" hinzufügen
df["season"] = df["Date"].dt.month.apply(get_season)

# Ausgabe der ersten Zeilen
df.head()
```

```python
#Hinzufügen der Spalte ob ein warter Tag ist

# Neue Spalte "warm_day" basierend auf der gefühlten Temperatur (apparent temperature)
df["warm_day"] = df["apparent_temperature_mean (°C)"].apply(lambda x: "Yes" if x >= 20 else "No")

# Ausgabe der ersten Zeilen
df.head()
```

```python
#Hinzufügen einer Spalte für rainy days

df["rainy_day"] = df["weathercode (wmo code)"].apply(lambda x: "Yes" if x > 50 else "No")
#df["rainy_day"] = df["weathercode (wmo code)"].apply(lambda x: "Ja" if x in [51, 53, 55, 61, 63, 65, 80, 81, 82, 95, 96, 99] else "Nein")
df.head()
```

```python
#Categorise the weather code

# transform into categorical type
df['season'] = pd.Categorical(df['season'], categories=['Spring', 'Summer', 'Fall','Winter'])
df['day_type'] = pd.Categorical(df['day_type'], categories=['Workday', 'Weekend'])          
```

```python
# Erstellen einer Wind-Kategorie für sinnvollere Gruppierung
def categorize_wind(speed):
    if speed < 10:
        return "Low Wind (<10 km/h)"
    elif speed < 20:
        return "Moderate Wind (10-20 km/h)"
    elif speed < 30:
        return "Strong Wind (20-30 km/h)"
    else:
        return "Very Strong Wind (>30 km/h)"

df["wind_category"] = df["windspeed_10m_max (km/h)"].apply(categorize_wind)

# 🔹 Richtige Kategorie-Namen setzen (genau wie in `categorize_wind()`)
wind_order = ["Low Wind (<10 km/h)", "Moderate Wind (10-20 km/h)", 
              "Strong Wind (20-30 km/h)", "Very Strong Wind (>30 km/h)"]

# 🔹 Kategorische Reihenfolge festlegen
df["wind_category"] = pd.Categorical(df["wind_category"], categories=wind_order, ordered=True)

df.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("CBS_2021-2023_Hourly_Weather.csv", sep =',')

df.head()
```

```python
#HinzufÃ¼gen der Spalte Werktage
import pandas as pd

df["Date"] = pd.to_datetime(df["Date"])

# PrÃ¼fen, ob der Tag ein Wochenende ist (Samstag=5, Sonntag=6)
df["is_weekend"] = df["Date"].dt.weekday >= 5  # True = Wochenende, False = Werktag

# Optional: Als Text anzeigen
df["day_type"] = df["is_weekend"].map({True: "Weekend", False: "Workday"})

df.head()
```

```python
#HinzufÃ¼gen der Spalte Holiday

import holidays
# Sicherstellen, dass die Date-Spalte datetime ist
df["Date"] = pd.to_datetime(df["Date"])

# US-Feiertage abrufen (Ã¤ndere "US" zu einem anderen LÃ¤ndercode, falls nÃ¶tig)
us_holidays = holidays.US()

# PrÃ¼fen, ob das Datum ein Feiertag ist
df["is_holiday"] = df["Date"].apply(lambda x: "Yes" if x in us_holidays else "No")
df.head()
```

```python
#HinzufÃ¼ger einer Spalte fÃ¼r Seasons

# Funktion zur Bestimmung der Jahreszeit basierend auf dem Monat
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Fall"

# Neue Spalte "season" hinzufÃ¼gen
df["season"] = df["Date"].dt.month.apply(get_season)

# Ausgabe der ersten Zeilen
df.head()
```

```python
# The correct column name for temperature is 'temperature_2m (Â°C)'
# Define a function to categorize temperature into 'Cold', 'Mild', and 'Warm'
def categorize_temperature(temp):
    if temp <= 10:
        return "Cold"
    elif 10 < temp <= 20:
        return "Mild"
    else:
        return "Warm"

# Apply the function to create a new 'Temperature Category' column
df["Temperature Category"] = df["temperature_2m (Â°C)"].apply(categorize_temperature)

df.head()
```

```python
# Define a function to categorize weather into 'Rainy Day', 'Snowy Day', and 'Dry Day' based on WMO weather codes
def categorize_weather(wmo_code):
    if wmo_code in [51, 53, 55, 61, 63, 65, 80, 81, 82]:  # Codes for rain
        return "Rainy Day"
    elif wmo_code in [71, 73, 75, 85, 86]:  # Codes for snow
        return "Snowy Day"
    else:
        return "Dry Day"

# Apply the function to create a new 'Weather Category' column
df["Weather Category"] = df["weathercode (wmo code)"].apply(categorize_weather)

df.head()
```

```python
# Erstellen einer Wind-Kategorie fÃ¼r sinnvollere Gruppierung
def categorize_wind(speed):
    if speed < 10:
        return "Low Wind (<10 km/h)"
    elif speed < 20:
        return "Moderate Wind (10-20 km/h)"
    elif speed < 30:
        return "Strong Wind (20-30 km/h)"
    else:
        return "Very Strong Wind (>30 km/h)"

df["wind_category"] = df["windspeed_10m (km/h)"].apply(categorize_wind)

# ðŸ”¹ Richtige Kategorie-Namen setzen (genau wie in `categorize_wind()`)
wind_order = ["Low Wind (<10 km/h)", "Moderate Wind (10-20 km/h)", 
              "Strong Wind (20-30 km/h)", "Very Strong Wind (>30 km/h)"]

# ðŸ”¹ Kategorische Reihenfolge festlegen
df["wind_category"] = pd.Categorical(df["wind_category"], categories=wind_order, ordered=True)

df.head()
```

```python
#Categorise the weather code

# transform into categorical type
df['season'] = pd.Categorical(df['season'], categories=['Spring', 'Summer', 'Fall','Winter'])
df['day_type'] = pd.Categorical(df['day_type'], categories=['Workday', 'Weekend'])  
df['Weather Category'] = pd.Categorical(df['Weather Category'], categories=['Dry Day', 'Rainy Day' , 'Snowy Day'])          
```

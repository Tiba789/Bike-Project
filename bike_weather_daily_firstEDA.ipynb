```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("CBS_2021-2023_Daily_Weather.csv", sep =',')

df.head()
```

```python
#Hinzufügen der Spalte Werktage
import pandas as pd

df["Date"] = pd.to_datetime(df["Date"])

# Prüfen, ob der Tag ein Wochenende ist (Samstag=5, Sonntag=6)
df["is_weekend"] = df["Date"].dt.weekday >= 5  # True = Wochenende, False = Werktag

# Optional: Als Text anzeigen
df["day_type"] = df["is_weekend"].map({True: "Wochenende", False: "Werktag"})

df.head()
```

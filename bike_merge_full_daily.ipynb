```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Laden der 5-Minuten-Daten
df_bike = pd.read_csv("CBS_2021-2023_Full.csv", sep =',')

# Konvertieren der Zeitspalte in ein Datetime-Objekt
df_bike['started_at'] = pd.to_datetime(df_bike['started_at'])

# Setzen der Zeitspalte als Index
df_bike.set_index('started_at', inplace=True)

# Nur numerische Spalten aggregieren
df_bike_numeric = df_bike.select_dtypes(include=[np.number])

# Aggregieren der 5-Minuten-Daten auf tägliche Daten
df_bike_daily = df_bike_numeric.resample('D').sum().reset_index()

# Laden der täglichen Wetterdaten
df_weather = pd.read_csv("CBS_2021-2023_Daily_Weather.csv", sep =',')

# Konvertieren der Zeitspalte in ein Datetime-Objekt
df_weather['Date'] = pd.to_datetime(df_weather['Date'])

# Zusammenführen der beiden DataFrames auf der Basis des Datums
df_merged = pd.merge(df_bike_daily, df_weather, left_on='started_at', right_on='Date')

# Speichern des zusammengeführten DataFrames in einer neuen CSV-Datei
df_merged.to_csv("CBS_2021-2023_Merged.csv", index=False)

# Anzeigen der ersten Zeilen des zusammengeführten DataFrames
df_merged.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Laden der 5-Minuten-Daten und Filtern der relevanten Spalten
df_bike = pd.read_csv("CBS_2021-2023_Full.csv", sep =',')
df_bike = df_bike[['started_at', 'start_station_name', 'end_station_name', 'start_station_id', 'end_station_id']]

# Konvertieren der Zeitspalte in ein Datetime-Objekt
df_bike['started_at'] = pd.to_datetime(df_bike['started_at'])

# Setzen der Zeitspalte als Index
df_bike.set_index('started_at', inplace=True)

# Aggregieren der Daten auf tägliche Basis und Zählen der Benutzungen jeder Station
df_bike_daily = df_bike.resample('D').apply(lambda x: x.value_counts()).reset_index()

# Umbenennen der Spalten für Klarheit
df_bike_daily.columns = ['date', 'station_name', 'station_id', 'count']

# Berechnen der Summe der Benutzungen jeder Station pro Tag
df_bike_daily_sum = df_bike_daily.groupby(['date', 'station_name', 'station_id']).sum().reset_index()

# Speichern des aggregierten DataFrames in einer neuen CSV-Datei
df_bike_daily_sum.to_csv("CBS_2021-2023_Station_Usage_Daily.csv", index=False)

# Anzeigen der ersten Zeilen des aggregierten DataFrames
df_bike_daily_sum.head()
```
